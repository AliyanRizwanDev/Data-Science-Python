```python
def palindrome(value):
    str_value = str(value)
    return str_value == str_value[::-1]


def check(input_set):
    result = []
    for value in input_set:
        if palindrome(value):
            result.append((value, "palindrome"))
        else:
            result.append((value, "symmetrical"))
    return result

userinput = set()
values = int(input("Enter the number of values in the set: "))
for i in range(values):
    value = input(f"Enter value {i + 1}: ")
    userinput.add(value)

output = check(userinput)

print("List:", output)
```

```python
class StudentManagementSystem:
    def __init__(self):
        self.students = {}
        self.courses = {
            "1": self.Course("1", "Data Science"),
            "2": self.Course("2", "English")
        }

    class Student:
        def __init__(self, studentId, name, age):
            self.studentId = studentId
            self.name = name
            self.age = age
            self.coursesEnrolled = {}
            self.grades = {}

        def enrollCourse(self, course):
            if course.courseId not in self.coursesEnrolled:
                self.coursesEnrolled[course.courseId] = None
                course.addStudent(self.studentId)

        def unenrollCourse(self, course):
            if course.courseId in self.coursesEnrolled:
                del self.coursesEnrolled[course.courseId]
                course.removeStudent(self.studentId)

        def updateGrade(self, courseId, grade):
            if courseId in self.coursesEnrolled:
                self.grades[courseId] = grade

        def displayDetails(self):
            print(f"Student ID: {self.studentId}")
            print(f"Name: {self.name}")
            print(f"Age: {self.age}")
            print("Courses Enrolled:")
            for courseId, grade in self.coursesEnrolled.items():
                print(f"  - Course ID: {courseId}, Grade: {grade if grade else 'N/A'}")

        def calculateAverageGrade(self):
            if not self.grades:
                return 0.0
            totalGrades = sum(self.grades.values())
            return totalGrades / len(self.grades)

    class Course:
        def __init__(self, courseId, courseName):
            self.courseId = courseId
            self.courseName = courseName
            self.studentsEnrolled = {}

        def addStudent(self, studentId):
            self.studentsEnrolled[studentId] = None

        def removeStudent(self, studentId):
            if studentId in self.studentsEnrolled:
                del self.studentsEnrolled[studentId]

        def listStudents(self):
            return list(self.studentsEnrolled.keys())

obj = StudentManagementSystem()

print("\nStudent Management System Menu:")
print("1. Add a new student")
print("2. Remove a student")
print("3. Update student grade")
print("4. Display student details")
print("5. Calculate average grade for a student")
print("6. List students enrolled in a course")
print("7. Enroll a student in a course")
print("8. Unenroll a student from a course")
print("0. Exit")

while True:
    choice = input("\nEnter your choice: ")

    if choice == "1":
        studentId = input("Enter student ID: ")
        name = input("Enter student name: ")
        age = input("Enter student age: ")
        student = obj.Student(studentId, name, int(age))
        obj.students[studentId] = student
        print("Student added successfully.")

    elif choice == "2":
        studentId = input("Enter student ID to remove: ")
        if studentId in obj.students:
            del obj.students[studentId]
            print("Student removed successfully.")
        else:
            print("Student not found.")

    elif choice == "3":
        studentId = input("Enter student ID to update grade: ")
        if studentId in obj.students:
            student = obj.students[studentId]
            courseId = input("Enter course ID: ")
            grade = float(input("Enter new grade: "))
            student.updateGrade(courseId, grade)
            print("Grade updated successfully.")
        else:
            print("Student not found.")

    elif choice == "4":
        studentId = input("Enter student ID to display details: ")
        if studentId in obj.students:
            student = obj.students[studentId]
            student.displayDetails()
        else:
            print("Student not found.")

    elif choice == "5":
        studentId = input("Enter student ID to calculate average grade: ")
        if studentId in obj.students:
            student = obj.students[studentId]
            averageGrade = student.calculateAverageGrade()
            print(f"Average Grade: {averageGrade:.2f}")
        else:
            print("Student not found.")

    elif choice == "6":
        courseId = input("Enter course ID: ")
        if courseId in obj.courses:
            course = obj.courses[courseId]
            enrolledStudents = course.listStudents()
            print(f"Students enrolled in course {courseId}: {enrolledStudents}")
        else:
            print("Course not found.")

    elif choice == "7":
        studentId = input("Enter student ID: ")
        courseId = input("Enter course ID: ")
        if studentId in obj.students and courseId in obj.courses:
            student = obj.students[studentId]
            course = obj.courses[courseId]
            student.enrollCourse(course)
            print("Student enrolled in course successfully.")
        else:
            print("Student or Course not found.")

    elif choice == "8":
        studentId = input("Enter student ID: ")
        courseId = input("Enter course ID: ")
        if studentId in obj.students and courseId in obj.courses:
            student = obj.students[studentId]
            course = obj.courses[courseId]
            student.unenrollCourse(course)
            print("Student unenrolled from course successfully.")
        else:
            print("Student or Course not found.")

    elif choice == "0":
        print("Exiting the system.")
        break
```
