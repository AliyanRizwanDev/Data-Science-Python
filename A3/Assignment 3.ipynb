```python
#EX1

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Assignment 3 Dataset.csv")

df['Total Profit'] = df.sum()

plt.plot(df['month_number'], df['total_profit'])

plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit for Each Month')

plt.show()
```

```python
# EX2
plt.plot(df['month_number'], df['total_profit'], linestyle='--', color='red', marker='o',label='Profit Data of Last Year')


plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.title('Total Profit with Style Properties')

plt.legend(loc='lower right')


plt.show()
```
