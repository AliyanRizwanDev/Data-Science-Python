```python
import numpy as np

arr = np.array([
[1,2,3,4,5],
[16,17,18,19,6],
[15,24,25,20,7],
[14,23,22,21,8],
[13,12,11,10,9]])


print("Original Array:\n")
for i in arr:
    print(i)

print('\n')


subarr=np.empty((0,3), dtype=int)
for i in range(1,4):
    temp = np.array([])
    for j in range(1,4):
        value = np.array([])
        value = arr[i][j]
        temp = np.append(temp,[value], axis=0)
    subarr = np.append(subarr,[temp], axis=0)


print("Sub Array:\n")  
for i in subarr:
    print(i)

print('\n')

reversearr = np.empty((0,5), dtype=int)
reverse = 4
for i in range(0,5):
    temp = np.array([])
    temp = arr[reverse]
    reversearr = np.append(reversearr,[temp], axis=0)
    reverse = reverse - 1

arr = np.copy(reversearr)

print("Array Reversed:\n")  
for i in arr:
    print(i)
    
print('\n')
print('The Indices of all elements greater than 15\n')

for i in range(0,5):
    for j in range(0,5):
        if arr[i][j] > 15:
            print("Element = {0}, Index = [{1}][{2}]".format(arr[i][j],i,j))
    
    
    
print('\n')
oneDarr = np.empty((0), dtype=int)
for i in range(0,5):
    for j in range(0,5):
        value = np.array([])
        value = arr[i][j]
        oneDarr = np.append(oneDarr,[value], axis=0)


    
arr = np.copy(oneDarr)
print("Array Converted to 1D:\n")  
for i in arr:
    print(i,end=" ")

print('\n')



```

```python
import numpy as np

    
def check(board):
    
    for row in range(0,10):
        
        if np.all(board[row] == 1) or np.all(board[row] == 2):
            
            winner = board[row, 0]
            return winner
        
    for col in range(0,10):
        
        if np.all(board[:, col] == 1) or np.all(board[:, col] == 2):
            
            winner = board[0, col]
            return winner


    if np.all(np.diag(board) == 1) or np.all(np.diag(board) == 2):
        
        winner = board[0, 0] 
        return winner
    
    if np.all(np.diag(np.fliplr(board)) == 1) or np.all(np.diag(np.fliplr(board)) == 2):
        winner = board[0, 9] 
        return winner
    
    if 0 in board:
        return -1
    else:
        return 0
    
    
    
    
    
board = np.zeros((10, 10), dtype=int)
board[2, :10] = 1  
board[5:, 5] = 2

print("Board\n")
print(board)

result = check(board)
if result == -1:
    print("\nGame in progress")
elif result == 0:
    print("\nTie")
else:
    print("\nWinner Player:",result)

```
